# Perceptron-Based Classification (Intro to DL)

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

from sklearn.linear_model import Perceptron
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

## About the MNIST Dataset
- MNIST (Modified National Institute of Standard and Technology)
- MNIST built a dataset of handwritten digits that is commonly used for training various image processing algorithms (DNN or CNN) 
- It's considered the "Hello, World" of deep learning
- The images are black and white and fit into 28x28 pixel bounding box 
- The dataset for this exercise is the image flattened. Therefore, every row represents an image.
- We have digits from 0 to 9
- Every pixel has a range between 0 to 255. 0 means absence of ink/color (dark) and 255 means full color/ink (white). Gray is half of 255

### Preprocessing

```python
X = df.drop(columns='label', axis=1)
y = df['label']

X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=10)
```

### Model Building

```python
# define the perceptron function
per_model = Perceptron()
```

```python
per_model.fit(X_train, y_train)
```

### Evaluation

```python
y_pred = per_model.predict(X_test)
y_pred
```

```python
accu_score = accuracy_score(y_test, y_pred)
```

```python
print('Accuracy Score:',accu_score)
```

Use the technique above 
1. Grab a sample of images (flattened data)
2. predict the digit
3. convert features into a 2D image
4. plot

```python
# select a sample of the digits
num_samples = 15

sample_X = X.sample(num_samples)
sample_X
```

```python
sample_labels_pred = per_model.predict(sample_X) 

#convert to 2D
sample_images = sample_X.values.reshape(num_samples, 28, 28) # we use 3 numbers since it's an array of images (img count, row, col)
```

```python
plt.figure(figsize=(28,3))
plt.suptitle('Predicted Label vs Actual Digits')
for img in range(num_samples):
    plt.subplot(1, num_samples, img +1)
    plt.imshow(sample_images[img], cmap='gray')
    plt.title(f'Pred: {sample_labels_pred[img]}')
    plt.axis('off')

```
