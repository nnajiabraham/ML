```python
df['Date'] = pd.to_datetime(df['Date'], format='%d/%m/%Y')
```

```python

# Create and add to Dataset Dt_Customer_Month for the Month portion of the Dt_Customer data
df['Dt_Month'] = df['Date'].dt.month
# Create and add to Dataset Dt_Customer_Day for the Day portion of the Dt_Customer data
df['Dt_Day'] = df['Date'].dt.day
# Create and add to Dataset Dt_Customer_Year for the Year portion of the Dt_Customer data
df['Dt_Year'] = df['Date'].dt.year
df.head()
```

```python
df['day_of_week_num'] = df['Date'].dt.weekday
df['weekday'] = df['day_of_week_num'].map(lambda d: 0 if d in [5,6] else 1)
df.head()
```

- columns dt_month, dt_day, dt_year were added as components of the date
- day_of_week_num has been added to represent each day of the week (0-6) monday: 0, Sunday :6
- weekday column has been added to represent weekdays (1) weekends (0)

```python
X = df.drop(columns=['Rented Bike Count', 'Date'])
y = df['Rented Bike Count']
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV

```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y, test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet

from sklearn.metrics import mean_squared_error, r2_score
```

```python
models = [
    {'name' : 'Linear Regression', 'model' : LinearRegression() },
    {'name' : 'Ridge Regression', 'model' : Ridge(), 'params' : {'max_iter': [5000],'alpha' : [0.01, 0.1, 1, 10, 100]} },
    {'name' : 'Lasso Regression', 'model' : Lasso(), 'params' : {'max_iter': [5000],'alpha' : [0.01, 0.1, 1, 10, 100]} },
]
```

```python
#track and store the parameters for a final evaluation
model_names = []
best_parameters = []
mean_sq_err_scores = []
r2_scores = []

#iterate over each model
for model_info in models:
    model_gs = GridSearchCV(estimator=model_info['model'], 
                            param_grid=model_info.get('params', {},), 
                            cv=5, 
                            )

    model_gs.fit(X_train, y_train)
    #additional evaluation with test data
    
    gs_best_model = model_gs.best_estimator_ #pick model with best hyperparametesr
    y_pred = gs_best_model.predict(X_test)
    #calculate the predicted values for evaluation
    mse_val = mean_squared_error(y_test, y_pred)
    r2_val = r2_score(y_test, y_pred)
    #append metadata to the empty lists
    model_names.append(model_info.get('name'))
    best_parameters.append(model_gs.best_params_)
    mean_sq_err_scores.append(mse_val)
    r2_scores.append(r2_val)
```

```python
result_dict = {'model_names' : model_names,
               'best_parameters' : best_parameters,
               'mean_sq_err_scores': mean_sq_err_scores,
               'r2' : r2_scores
               }
```
