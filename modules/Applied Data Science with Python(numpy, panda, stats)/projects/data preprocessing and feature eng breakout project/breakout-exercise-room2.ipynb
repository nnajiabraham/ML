### 2. Data Cleansing

a) Missing Value Handling

```python
df.isna().sum()
```

No missing values

b) Outlier Detection

```python
fig, axes = plt.subplots(nrows=6, ncols=2, figsize=(14,16))

#flatten the axes so we can have a series from 0 to 3 instead of [0,0] for example
axes = axes.flatten()

fig.suptitle("Distribution of features")

#loop through each column name and plot it
for idx, col in enumerate(numerical_cols):
    sns.boxplot(data=df, x=col, ax=axes[idx])

fig.tight_layout(pad=1.2, w_pad=1)
```

The columns with the highest number of outliers are fixed acidity, volatile acidity and chlorides

```python
df[['fixed acidity', 'volatile acidity', 'chlorides']].describe()
```

```python
highest_outlier_cols = ['fixed acidity', 'volatile acidity', 'chlorides']

for col in highest_outlier_cols:
    q1 = np.quantile(df[col], .25)
    q3 = np.quantile(df[col], .75)

    iqr = q3 - q1

    upper_range = q3 + 1.5 * iqr
    lower_range = q1 - 1.5 * iqr

    df = df[(df[col] < upper_range) & (df[col] > lower_range)]

df
```

```python
fig, axes = plt.subplots(nrows=6, ncols=2, figsize=(14,16))

#flatten the axes so we can have a series from 0 to 3 instead of [0,0] for example
axes = axes.flatten()

fig.suptitle("Distribution of features")

#loop through each column name and plot it
for idx, col in enumerate(numerical_cols):
    sns.boxplot(data=df, x=col, ax=axes[idx])

fig.tight_layout(pad=1.2, w_pad=1)
```

### 3. Feature Engineering

a) Create New Features

```python
df['total_acidity'] = df['fixed acidity'] + df['volatile acidity']
df.head()
```

```python
df['sugar_to_alcohol_ratio'] = df['residual sugar'] / df['alcohol']
df
```

```python
df['density'].value_counts(bins=3)
```

```python
df['density_level'] = pd.cut(df['density'], 3, labels=[1,2,3])
df
```

```python
df['sugar_level'] = pd.cut(df['residual sugar'], 3, labels=[1,2,3])
df
```

```python
df['sugar_level'].value_counts()
```
