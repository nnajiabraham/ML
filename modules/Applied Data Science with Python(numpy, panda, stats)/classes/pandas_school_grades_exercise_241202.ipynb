Here are the different subtypes you can use:

- int8 / uint8 : consumes 1 byte of memory, range between -128/127 or 0/255
- bool : consumes 1 byte, true or false
- float16 / int16 / uint16: consumes 2 bytes of memory, range between -32768 and 32767 or 0/65535
- float32 / int32 / uint32 : consumes 4 bytes of memory, range between -2147483648 and 2147483647
- float64 / int64 / uint64: consumes 8 bytes of memory

````python
df.info(memory_usage='deep')
````

### Advanced Aggregation

use a custom function for aggregation

````python
def score_range(col):
    return col.max() - col.min()

df.groupby(['class_group'])['reading_score'].agg(['min', 'max', score_range])
````

````python
df.groupby(['class_group'])[df_num_cols].agg(['min', 'max', score_range])
````

````python
df.groupby(['gender','class_group'])[df_num_cols].agg(['min', 'max', score_range])
````

What if we want a different agg function for each numeric value

````python
df.groupby(['gender','class_group'])[df_num_cols].agg({'math_score':'mean',
                                                       'reading_score':'max',
                                                       'writing_score':'std'})
````

````python
df.groupby(['gender','class_group'])[df_num_cols].agg(avg_math_score=('math_score', 'mean'),
                                                      max_reading_score=('reading_score', 'max'))
````
